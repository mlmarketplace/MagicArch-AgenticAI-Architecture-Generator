## The Perceive → Decide → Act → Learn Cycle

### Perceive
• Gather data from sensors,
databases, user inputs or
digital environments

• Process and filter information
for relevance and priority

### Decide
• Analyse current state
against goals and
constraints

• Plan next steps based on
available options and
predicted outcomes

### Act
• Execute tasks or
interventions autonomously

• Implement decisions
through digital or physical
means

### Learn
• Evaluate results against
expectations

• Update knowledge base
and improve future
decisions

## Autonomy and Adaptability in Action

Agentic AI stands out by continuously
monitoring its environment and using
feedback loops to guide decisions.

Unlike traditional systems, agentic AI
can independently:
• Detect environmental changes
affecting performance
• Evaluate current strategy effectiveness
• Adjust approaches without human
reprogramming
• Optimise for short-term outcomes and
long-term goals

## What Does Autonomy Mean

- Autonomy is the
ability to act
independently
without step-by-
step instructions

- In AI, it reflects
how freely a
system chooses
actions to meet
objectives

#### Definition:
A system is autonomous if, once instructed, it independenctly performs tasks and determines next steps basded on its environemnt and goals.

### Decide
- Access options and select the best path based on conditions and goals
### Plan
- Outline stpes to reach the outcome, factoring in obstacles and resources
### Act
Carry out the plan, adapt in real-time and learn each experience

## Tool 2: judge_output

### Purpose:
Score the generated output using LLM-as-judge and enforce pass/fail.

This tool never edits content. It only evaluates.

```python
import json
from langchain_core.messages import SystemMessage, HumanMessage


def judge_output(model_client, generated_text: str) -> dict:
    """
    Evaluates generated output using a strict LLM-as-judge rubric.
    """

    system_prompt = (
        "You are an uncompromising AI architecture reviewer.\n\n"
        "Your job is to judge whether the generated output demonstrates:\n"
        "- Architectural clarity\n"
        "- Decision discipline\n"
        "- Production realism\n\n"
        "You do not reward verbosity or neutrality.\n"
        "Penalize vague language, hedging, architectural hype, "
        "or multi-agent evangelism without justification."
    )

    judge_prompt = (
        "Score the response below on a scale of 0 to 5 for each dimension:\n\n"
        "1. Conceptual Clarity\n"
        "2. Decision Sharpness\n"
        "3. Example Quality\n"
        "4. Engineering Realism\n"
        "5. Anti-Hype Discipline\n\n"
        "Rules:\n"
        "- Hard fail if any score is 0\n"
        "- Pass only if total score >= 20/25\n"
        "- Output JSON only, with this schema:\n\n"
        "{\n"
        "  \"scores\": {\n"
        "    \"conceptual_clarity\": int,\n"
        "    \"decision_sharpness\": int,\n"
        "    \"example_quality\": int,\n"
        "    \"engineering_realism\": int,\n"
        "    \"anti_hype_discipline\": int\n"
        "  },\n"
        "  \"total_score\": int,\n"
        "  \"verdict\": \"PASS\" | \"FAIL\",\n"
        "  \"lowest_scoring_dimension\": str,\n"
        "  \"justification\": str\n"
        "}\n\n"
        "Response to evaluate:\n"
        f"{generated_text}"
    )

    messages = [
        SystemMessage(content=system_prompt),
        HumanMessage(content=judge_prompt)
    ]

    response = model_client.invoke(messages)

    return json.loads(response.content)
```
